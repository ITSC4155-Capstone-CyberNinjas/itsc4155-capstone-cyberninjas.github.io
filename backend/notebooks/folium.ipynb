```python
def prep_data(df):
    
    #read buildings file
    buildings = pd.read_csv('csv/buildings.csv').dropna()
    
    # separate columns of interest into lists
    buildings_list = list(buildings['Building abbreviation'])
    lat_list = list(buildings.Latitude)
    long_list = list(buildings.Longitude)

    lat_long_dict = { k:v for k,v in zip(buildings_list, zip(lat_list, long_list)) }

    # convert dictionary values to list (instead of tuple)
    for k in lat_long_dict.keys():
        lat_long_dict[k] = list(lat_long_dict[k])

    # list of timestamps
    ts_idx = [str(x) for x in pd.to_datetime(day.timestamp)]
    
    #drop building with nulls from data
    curr_cols = set(day.columns)
    to_keep = set(buildings_list)
    to_drop = list(curr_cols.difference(to_keep))
    day.drop(columns = to_drop, inplace = True)
    
    # normalize data 
    min_max_scaler = p.MinMaxScaler()
    norm_values = min_max_scaler.fit_transform(day.values)
    counts_norm = norm_values.tolist()
    
    # create data matrix for map
    data = []
    for row in counts_norm:
        new_row = []
        for idx, col in enumerate(day.columns):
            curr_coordinate = lat_long_dict[col].copy()
            curr_coordinate.append( row[idx] )
            new_row.append( 
                curr_coordinate
            )
        data.append(new_row)
        
    return data, ts_idx
        
```

```python
df = pd.read_csv('csv/wifi_counts_combined.csv')
day = df[72:96]
```

```python
data, idx = prep_data(day)
```

```python
OurMap = folium.Map(location=[35.30742034864125, -80.73590951022354], tiles="OpenStreetMap", zoom_start=15)

HeatMapWithTime(
    data, 
    index = idx,
    radius = 35
).add_to(OurMap)

OurMap
```
